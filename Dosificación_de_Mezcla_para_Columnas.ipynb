# **Planteamiento:**

Un ingeniero residente necesita calcular cuántos sacos de cemento se requieren para vaciar una columna de sección circular.

Se sabe que para un concreto estándar de 210 kg/cm2 (3000 psi), se consumen 9.5 sacos de cemento por cada metro cúbico de mezcla.

**Requerimientos:**

1. Solicitar la altura de la columna en metros.
2. Solicitar el diámetro de la columna en metros.
3. Calcular el volumen de la columna: $$V = π * r^2 * h$$

(Sugerencia: usar 3.1416
 o importar math.pi).

4. Calcular la cantidad de sacos: $$Sacos = V * 9.5$$

**Prueba del programa:**

Para una columna de 3 metros de alto y 0.4 metros de diámetro (r=0.2
):

Volumen ≈0.377 m3
.
Sacos necesarios ≈3.58

 (el estudiante verá aquí la importancia de los tipos decimales).

````python
# Entrada:Solicitar la altura de la columna en metros.
#Solicitar el diametro de la columna en metros.
h = float(input("Ingrese la altura de la columna en metros: "))
d = float(input("Ingrese el diamtero de la columna en metros: "))

# PROCESO: Calcula el radio r, calcula el volumen de la columna, calcula sacos
r = d/2
v = 3.1416 * r**2 * h
Sacos = v * 9.5

# SALIDA: Imprimir el resultado
print(f"Volumen ≈{v: .3f} m3")
print(f"Sacos necesarios ≈{Sacos: .2f}")
````

```markdown
El código para calcular los sacos de cemento funciona de la siguiente manera:

*   `h = float(input("Ingrese la altura de la columna en metros: "))`
    *   Esta línea solicita al usuario que ingrese la altura de la columna en metros. El valor ingresado se convierte a un número decimal (`float`) y se almacena en la variable `h`.

*   `d = float(input("Ingrese el diamtero de la columna en metros: "))`
    *   Similar a la anterior, esta línea solicita al usuario el diámetro de la columna en metros. El valor se convierte a `float` y se guarda en la variable `d`.

*   `r = d/2`
    *   Calcula el radio de la columna dividiendo el diámetro (`d`) por 2 y almacena el resultado en la variable `r`.

*   `v = 3.1416 * r**2 * h`
    *   Calcula el volumen de la columna utilizando la fórmula para el volumen de un cilindro (`V = π * r^2 * h`). Se usa `3.1416` como una aproximación de Pi, `r**2` para el radio al cuadrado, y `h` para la altura. El resultado se guarda en la variable `v`.

*   `Sacos = v * 9.5`
    *   Calcula la cantidad total de sacos de cemento necesarios multiplicando el volumen de la columna (`v`) por la cantidad de sacos requeridos por metro cúbico (`9.5`). El resultado se almacena en la variable `Sacos`.

*   `print(f"Volumen ≈{v: .3f} m3")`
    *   Imprime el volumen calculado de la columna. `f"..."` es una f-string que permite insertar variables directamente en el texto. `: .3f` formatea el volumen para mostrarlo con 3 decimales.

*   `print(f"Sacos necesarios ≈{Sacos: .2f}")`
    *   Imprime la cantidad calculada de sacos de cemento. `: .2f` formatea la cantidad de sacos para mostrarla con 2 decimales.
```
