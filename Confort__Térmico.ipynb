#**Métodos y Cálculos**

+ Índice de temperatura y humedad (producción intensiva)
$$ITH = 0.72 * (T_{bs} + T_{bs}) + 40.6$$

Donde:

1. Tbs: Temperatura bulbo seco (°C)
2. Tbh: Temperatura de bulbo húmedo (°C)

Para la correlación de los datos de la especie bovina en la producción de leche se recomienda
utilizar los límites propuestos por National Weather Service:

1. Sí $ITH ≤ 70$ ambiente no estresante
2. Sí $ITH > 70 \space and \space <= 78$ situación de alerta por estrés térmico
3. Sí $ITH > 78 \space and \space <= 83$ peligro por estrés térmico
4. Sí $ITH > 83$ situación crítica

````python
#ENTRADA: pedir tbs(temperatura de bulbo seco) y tbh (temperatura de bulbo húmedo)
tbs = float(input("Ingrese la temperatura de bulbo seco, grados centigrados: "))
tbh = float(input("Ingrese la temperatura de bulbo húmedo, grados centigrados: "))

#PROCESO: cálculo de ith (indice de temperatura y humedad)
ith = 0.72 * (tbs + tbh) + 40.6

#SALIDA: imprimir resultado, de la evaluacion del ith y el confort térmico
if ith <= 70:
  print("El ambiente es no estresante")
elif ith > 70 and ith <= 78:
  print("La situación es de alerta por estrés térmico")
elif ith > 78 and ith <= 83:
  print("La situación es de peligro por estrés térmico")
else:
  print("La situación es crítica")
````

### **Teoría del Confort Térmico en Ganado Vacuno**

El confort térmico en el ganado vacuno se refiere al estado en el que los animales no experimentan estrés por calor ni por frío, lo que les permite mantener su temperatura corporal central dentro de rangos óptimos sin un gasto excesivo de energía. Este equilibrio es crucial para la salud, bienestar y productividad de los animales, incluyendo la producción de leche y carne.

#### **Importancia del Confort Térmico:**

*   **Salud y Bienestar:** El estrés térmico puede debilitar el sistema inmunológico, haciendo a los animales más susceptibles a enfermedades.
*   **Productividad:** Las temperaturas fuera del rango de confort afectan negativamente la ingesta de alimento, la tasa de crecimiento, la producción de leche, la eficiencia reproductiva y la calidad del producto final.
*   **Eficiencia Alimenticia:** Los animales estresados térmicamente desvían energía para termorregulación en lugar de producción.

#### **Factores que Afectan el Confort Térmico:**

El confort térmico de los bovinos es influenciado por una combinación de factores ambientales y características del animal:

1.  **Temperatura del Aire (Tbs):** La temperatura ambiente es un factor principal. Temperaturas muy altas o muy bajas pueden causar estrés.
2.  **Humedad Relativa (HR):** Una alta humedad reduce la capacidad de los animales para disipar calor por evaporación (sudoración y jadeo), haciendo que temperaturas elevadas sean más estresantes.
3.  **Radiación Solar:** La exposición directa al sol aumenta significativamente la carga de calor en los animales.
4.  **Velocidad del Viento:** El viento puede ayudar a disipar calor en ambientes cálidos (enfriamiento convectivo) o aumentar la sensación de frío en ambientes fríos.
5.  **Precipitación:** La lluvia puede contribuir al enfriamiento, pero una combinación de frío y humedad puede ser muy estresante.
6.  **Características del Animal:** La raza (razas europeas vs. razas cebuinas, por ejemplo), el color del pelaje, la edad, el estado fisiológico (lactancia, gestación) y el nivel de producción influyen en la tolerancia al estrés térmico.

#### **Índices de Confort Térmico:**

Para cuantificar y monitorear el estrés térmico, se utilizan diversos índices que combinan múltiples factores meteorológicos en un solo valor. El **Índice de Temperatura y Humedad (ITH)** es uno de los más comunes y se utiliza ampliamente en la evaluación del estrés por calor en ganado vacuno. Como se mencionó anteriormente, este índice considera la temperatura de bulbo seco (Tbs) y la temperatura de bulbo húmedo (Tbh) para estimar la carga térmica que experimentan los animales y clasificar el nivel de estrés.
