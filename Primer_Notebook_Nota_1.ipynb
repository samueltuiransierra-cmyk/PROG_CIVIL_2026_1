# 1. Problema de ingeniería (ejemplo sencillo)

Un ingeniero agrícola necesita calcular **el área de un terreno rectangular** para estimar la cantidad de fertilizante que debe aplicar.

La fórmula es:

[
Área = largo \times ancho
]

---

# 2. Definición del algoritmo

Primero debemos identificar **Entradas, Proceso y Salidas**.

### Entradas (datos que recibe el programa)

* `largo` del terreno (metros)
* `ancho` del terreno (metros)

### Proceso (cálculo que realiza el algoritmo)

* Multiplicar largo por ancho

[
area = largo \times ancho
]

### Salida (resultado que entrega el programa)

* Área del terreno en **metros cuadrados**

---

# 3. Pseudocódigo

El pseudocódigo permite expresar la lógica **antes de escribir Python**.

```
Inicio

Leer largo
Leer ancho

Mostrar area

Fin
```

---

# 4. Implementación en Python

```python
# Programa para calcular el área de un terreno rectangular
# Aplicación básica en ingeniería agrícola

# Entrada de datos
largo = float(input("Ingrese el largo del terreno en metros: "))
ancho = float(input("Ingrese el ancho del terreno en metros: "))

# Proceso

# Salida
print("El área del terreno es:", area, "metros cuadrados")
```

---

# 5. Ejemplo de ejecución

```

```

```

```

### Un ejemplo aplicado a la Ingeniería

En lugar de un simple "Hola Mundo", veamos algo que podrías aplicar en **Google Colab**. Supongamos que necesitamos calcular el volumen de agua necesario para llenar un tanque cilíndrico de almacenamiento.

Antes de mostrarte el código, analicemos la lógica:

1. Necesitamos el **radio** y la **altura**.
2. Usamos la fórmula: $V = \pi \cdot r^2 \cdot h$.
3. Mostramos el resultado con unidades claras.

Aquí tienes un ejemplo de cómo estructuraríamos esto siguiendo **buenas prácticas**:

````python
import math

def calcular_volumen_tanque(radio_m, altura_m):
    """
    Calcula el volumen de un tanque cilíndrico en metros cúbicos.
    """
    volumen = math.pi * (radio_m ** 2) * altura_m
    return volumen

# Datos de entrada (Simulando una medición en campo)
radio_tanque = 2.5  # metros
altura_tanque = 5.0 # metros

# Llamada a la función
resultado_m3 = calcular_volumen_tanque(radio_tanque, altura_tanque)

print(f"El volumen del tanque es: {resultado_m3:.2f} m³")
````

### ¿Qué notarás en este código?

* **Nombres descriptivos:** No usamos `x` o `y`, sino `radio_m` para saber qué mide y en qué unidad está.
* **Modularización:** Usamos una función (`def`) para que el cálculo sea reutilizable.
* **Precisión:** Importamos el módulo `math` para usar el valor exacto de $\pi$.
